#### Plotting using penguins data

```python
# load sns dataset
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
# plot displot
sns.displot(data=penguins, x='flipper_length_mm', kde=True)
```

```python
sns.displot(data=penguins, x='flipper_length_mm', y='bill_length_mm')
```

```python
sns.displot(data=penguins, x='flipper_length_mm', y='bill_length_mm', kind='kde', rug=True)
```

```python
sns.displot(data=penguins, x='flipper_length_mm', hue='species', kind='kde')
```

```python
sns.displot(data=penguins, x='flipper_length_mm', hue='species', col='sex', kind='kde')
```

```python
sns.displot(data=penguins, x='flipper_length_mm', col='species', hue='sex', kind='kde')
```

```python
sns.displot(data=penguins, x='flipper_length_mm', hue='sex', col='species', kind='ecdf')
```

```python
s = sns.displot(data=penguins, y='flipper_length_mm', hue='sex', col='species', kind='ecdf')
s.set_axis_labels('Density (a.u)', 'Flipper length (mm)')
s.set_titles("{col_name} penguins")
```

#### Plottings flights dataset

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
# plotting line using may flights
may_flights = flights.query("month == 'May'")
sns.lineplot(data=may_flights, x='year', y='passengers')
```

```python
# pivoting
flights_wide = flights.pivot(index='year', columns='month', values='passengers')
flights_wide.head()
```

```python
# to plot specifice month data
sns.lineplot(data=flights_wide['May'])
```

```python
sns.lineplot(data=flights_wide)
```

```python
sns.lineplot(data=flights, x='year', y='passengers')
```
