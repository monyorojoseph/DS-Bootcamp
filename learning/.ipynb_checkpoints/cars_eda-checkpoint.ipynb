```python
# dropping irrelevant columns
cars = cars.drop(['Engine Fuel Type', 'Market Category', 'Number of Doors', 
                  'Popularity', 'Vehicle Style', 'Vehicle Size'], axis=1)
cars.head()
```

```python
# renaming columns
cars = cars.rename(columns={
    "Engine HP": "Hp",
    "Engine Cylinders": "Cylinders",
    "Transmission Type": "Transmission",
    "Driven_Wheels": "Drive Mode",
    "highway MPG": "MPG-H",
    "city mpg": "MPG-C",
    "MSRP": "Price"
    
})
cars.head()
```

```python
# dropping duplicate rows
duplicated_rows_df = cars[cars.duplicated()]
duplicated_rows_df.shape
```

```python
cars=cars.drop_duplicates()
cars.shape
```

```python
# dropping missing or null values
cars = cars.dropna()
cars.shape
```

```python
q1 = cars.quantile(0.25, numeric_only=True)
q3 = cars.quantile(0.75, numeric_only=True)
IQR = q3 - q1
print(IQR)
```

```python
cars = cars[~((cars < (q1-1.5 * IQR)) | (( cars > (q3+1.5 * IQR)))).any(axis=1)]
cars.shape
```
